### 122A Lucky Division - 1000

https://codeforces.com/problemset/problem/122/A

```python
def luckyNumbers(x):
    result = []
    for i in range(1, x+1):
        luckyNumber = True
        iString = str(i)
        for j in iString:
            if j not in '47':
                luckyNumber = False
        if luckyNumber:
            result += [i]    
    return result

numberToCheck = int(input())
luckyNubersList = luckyNumbers(numberToCheck)
almostLucky = False
for i in luckyNubersList:
    if numberToCheck % i == 0:
        almostLucky = True
        break

if almostLucky:
    print("YES")
else:
    print("NO")
```

### 102B Sum of Digits - 1000

https://codeforces.com/problemset/problem/102/B

```python
def addNumbers(myNumber):
    numberDigits = []
    numberString = str(myNumber)
    for i in numberString:
        numberDigits.append(int(i))
    return sum(numberDigits)

number = input()
myNumber = int(number)
magicPerformed = 0
while myNumber > 9:
    myNumber = addNumbers(myNumber)
    magicPerformed += 1
print(magicPerformed)
```

### 239A Two Bags of Potatoes - 1200

https://codeforces.com/problemset/problem/239/A

```python
y, k, n = input().split(" ")
secondBag = int(y)
divisibleBy = int(k)
potatoTotalMax = int(n)
firstBagPosibilities = []
for i in range(0, (potatoTotalMax+1), divisibleBy):
    if i > secondBag:
        firstBagPosibilities += [(i - secondBag)]
        
if len(firstBagPosibilities) == 0:
    print(-1)
else:
    for i in range(len(firstBagPosibilities)):
        print(firstBagPosibilities[i], end=" ")
```

### 25A IQ Test - 1300

https://codeforces.com/problemset/problem/25/A

```python
def isEven(x):
    isEven = False
    if int(x) % 2 == 0:
        isEven = True
    return isEven

testCount = int((input().split())[0])
testNumbers = input().split(" ")
evensCount = 0

#Eerste 3 getallen nakijken
for i in range(3):
    if isEven(testNumbers[i]):
        evensCount += 1

if evensCount >= 2:
    for j in range(testCount+1):
        if isEven(testNumbers[j]) == False:
            print(j+1)
            break
else:
    for j in range(testCount+1):
        if isEven(testNumbers[j]) == True:
            print(j+1)
            break
```
