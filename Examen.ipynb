```python
def islucky(n):
    flag = 1
    for ch in str(n):
        if '4' == ch or '7' == ch:
            continue
        else:
            flag = 0
    return flag


def isAlmostLucky(n):
    luckyNumbers = []

    for i in range(1000):
        if islucky(i):
            luckyNumbers.append(i)
    for ln in luckyNumbers:
        if n % ln == 0:
            return "YES" 
    return "NO"

test = int(input())

val = isAlmostLucky(test)
print(val)
```

### 230B T-primes - 1300

https://codeforces.com/problemset/problem/230/B

```python
from math import sqrt

max_val = 1000000

def primes_sieve():
    primes = [True] * (max_val + 1)  
    
    primes[0] = False  
    primes[1] = False  
    
    for p in range(2, int(sqrt(max_val)) + 1):
        if primes[p]:  
            for i in range(p * p, max_val + 1, p):
                primes[i] = False
    
    return primes 

def checkSqrtIsPrime(n):
    if n == 4:  
        return True
    if n < 4 or n % 2 == 0:  
        return False
    sqrt_n = n**0.5  
    if sqrt_n == int(sqrt_n):  
        if primes[int(sqrt_n)] == True:
            return True
    return False 


inp = int(input())  
lst = list(map(int, input().split())) 

primes = primes_sieve()


for i in lst:
    if checkSqrtIsPrime(i):  
        print("YES")  
    else:
        print("NO") 
```

### 1472B Fair Division - 800

https://codeforces.com/problemset/problem/1472/B

```python
def can_divide_candies(test_cases):
    results = []
    
    for n, weights in test_cases:
        total_weight = sum(weights)
        
        if total_weight % 2 != 0:
            results.append("NO")
            continue
        
        target_weight = total_weight // 2
        count_1 = weights.count(1)
        count_2 = weights.count(2)
 
        max_2s_weight = min(target_weight // 2, count_2) * 2
        remaining_weight = target_weight - max_2s_weight
        
        if remaining_weight <= count_1:
            results.append("YES")
        else:
            results.append("NO")
    
    return results
 

t = int(input())
test_cases = []
 
for _ in range(t):  
    n = int(input())
    weights = list(map(int, input().split()))
    test_cases.append((n, weights))
 

results = can_divide_candies(test_cases)
 

for result in results:
    print(result)
```

### 2050C Uninteresting Number

https://codeforces.com/problemset/problem/2050/C

```python
t = int(input())
testcases = [input() for case in range(t)]
def divisible_by_9(digit):
    count2=count3=som = 0
    for i in digit:
        if i == '2':
            count2+=1
        elif i == '3':
            count3+=1
        som+=int(i)
        som%=9
    for i in range(count3+1):
        for j in range(count2+1):
            temp = som
            temp = som + (j*2+i*6)
            if temp%9==0:
                return 'YES'
    return 'NO'
 
for case in testcases:
    print(divisible_by_9(case))
```
